# Condicional

```python
# If: se acontecer algo
if 5 < 10:
  print('Verdade')  # Tudo que colocar neste nível do espaçamento, vai representar essa opção
                    # Se fosse falso, ele não executaria essa parte do comando
```

```python
# If/else
num1 = 10
num2 = 5
if num1 > num2:
  print('O num1 é o maior número')
else:
  print('O num2 é o maior número')
```

```python
# If / Elif / Else   # If: pra testar uma verdade; # Else: se é verdade ou falso; # Elif: pra quando quer fazer vários testes
num1 = 10            # Pode turbinar com os operadores lógicos pra fazer várias comparações nas condicionais
num2 = 10
if num1 > num2:
  print('O num1 é o maior número')
elif num1 < num2:
  print('O num2 é o maior número')
else:
  print('Os dois são iguais')
```

# Condicional com operadores lógicos

```python
num1 = 10
num2 = 5
num3 = 2
if (num1 > num2) and (num1 > num3):
  print('O num1 é o maior número')
else:
  print('O num1 não é o maior número')
```

```python
num1 = 10     # Teste
num2 = 5
num3 = 20
if (num1 > num2) and (num1 > num3):
  print('O num1 é o maior número')
elif (num2 > num1) and (num2 > num3):
  print('O num2 é o maior número')
else:
  print('O num3 é o maior número')
```
