# Vizualização de Variáveis e Textos

```python
print('conhecendo o comando Print')

x = 8
print(x)   # Não pode colocar o x entre aspas, pois se não, o comando vai entender que x é um texto e não uma variável.
```

```python
y = 10    # Soma de variáveis
soma = x + y
print(soma)
```

```python
nome = 'Thaynnara Alves'   # Variáveis do tipo string
print(nome)
```

**Concatenação**: mostra as variáveis junto com os textos. Usa-se o operador de *vírgula*.

```python
x = 5
print('O valor de x é', x)
```

```python
print('O valor de x é', x, 'e o dobro é', x * 2)        # Operações aritméticas

# Evitar colocar expressões longas, nesse caso, faz mais sentido colocar dentro de uma variável, do que escrever ela toda.
```

```python
print('Eu,', nome, 'tirei', x, 'na prova de Python')
```

```python
y = 10; z = 20 + x - y   # Mais de uma variável separando com vírgulas
# Quando estiver declarando variáveis, separar com ponto e vírgula outras variáveis na mesma linha

print('Declarar várias variáveis na mesma linha:', y, z)
```

Type

```python
x = 10
type(x)
print('x', type(x))   # Neste caso, não precisaria usar o print, basta o type

nome = 'Thaynnara'    # String sempre será reconhecido pelas aspas
print('nome', type(nome))

num = '100'
num_convert = int(num)  # Converter variável
print('num', type(num))
print('num_convert', type(num_convert))

num_float = float(num)    # O ponto indica que é um numero real
print('num_float', num_float)
```

Input
* Armazenar a informação digitada pelo usuário em uma variável
* O dado armazenado sempre será um texto

```python
nome = input('Digite o nome do usuário ')     # Fica aguardando o usuário digitar algo
print(nome)
```

```python
valor = input('Digite a nota ')

print(nome, 'sua nota foi ', valor)
```

```python
valor = input('Digite o valor do produto: ')

print(valor)
```

```python
desconto = valor - 5   # Código errado, pois ele não consegue fazer uma operação de subtração em um campo que é do tipo string, precisaria ser do tipo int
```

```python
valor_convert = int(valor)    # Utiliza-se essa operação quando o valor convert for utilizado em outros lugares do seu código
desconto = valor_convert - 5
print(desconto)
```

```python
desconto = int(valor) - 5    # Utiliza-se essa operação quando o valor convert for utilizado somente nesta linha no programa inteiro
print(desconto)              # Operação direta
```

Exemplo de código executável (em Python) que ilustra boas práticas:

```python
# Solicita o nome do usuário e o exibe
nome_usuario = input("Por favor, insira seu nome: ")
print(f"Olá, {nome_usuario}! Bem-vindo ao nosso sistema.")
```

```python
# Solicita a idade do usuário e converte para inteiro
idade_usuario = int(input("Por favor, insira sua idade: "))
```

```python
# Verifica se o usuário é maior de idade
if idade_usuario >= 18:
    print("Você é maior de idade.")
else:
    print("Você é menor de idade.")
```
